```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint

#cat1 = "shop-laptops"
cat2 = "shop-tvs"
url = "https://www.lazada.sg/"+cat2
base_url=url
url_list = []

def get_and_convert_to_soup(url):
    content = requests.get(url).text
    soup = BeautifulSoup(content, "html.parser")
    return soup


def go_to_a_page_and_collect_p_details_url(url):
    global base_url
    soup = get_and_convert_to_soup(url)
    products_details_url_list =  soup.find_all("a" , {"class":"c-product-card__img-placeholder-inner"} )
    tmp_list = []
    for p in products_details_url_list:
        product_url = p.get("href")
        full_url = base_url + product_url
        tmp_list.append( full_url )
        print full_url
    return tmp_list

list_of_products_details_url = go_to_a_page_and_collect_p_details_url(url)
url_list.extend( list_of_products_details_url )



def get_product_details(url):
    #prod_title
    soup = get_and_convert_to_soup(url)
    h1_tag = soup.find("h1", {"id":"prod_title"})
    print (h1_tag.text.strip())
    
    s_price_tag = soup.find("span", {"id":"special_price_box"})
    print (s_price_tag.text.strip())

for page_number in range(2,3):
    page_url = url+"?page=%s"%page_number
    #print (page_url)
    list_of_products_details_url = go_to_a_page_and_collect_p_details_url(page_url)
    url_list.extend( list_of_products_details_url )
    
    
for url in url_list:
    get_product_details(url)
```
