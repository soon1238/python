# Iris

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

iris_data = pd.read_csv('iris-data.csv')
iris_data.head()
```

```python
iris_data.describe()
```

```python
from matplotlib import pyplot as plt
iris_data.plot()
plt.show()
```

```python
iris_data.groupby('class').nunique()
```

```python
iris_data.loc[iris_data['class'] == 'versicolor', 'class'] = 'Iris-versicolor'
iris_data.loc[iris_data['class'] == 'Iris-setossa', 'class'] = 'Iris-setosa'

iris_data.groupby('class').nunique()
```

```python
iris_data['class_code'] = pd.Categorical(iris_data['class'])
codes = iris_data['class_code'].cat.codes

plt.scatter(iris_data['sepal_length_cm'],codes,c=codes, cmap='prism')
plt.show()
```

```python
plt.scatter(iris_data['sepal_width_cm'], codes,c=codes, cmap='prism')
plt.show()
```

```python
iris_data['sepal_width_cm'].hist(bins=20)
plt.show()
```

```python
plt.scatter(iris_data['petal_width_cm'], codes,c=codes, cmap='prism')
plt.show()
```

```python
plt.scatter(iris_data['petal_length_cm'], codes,c=codes, cmap='prism')
plt.show()
```

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull()) |
              (iris_data['sepal_width_cm'].isnull()) |
              (iris_data['petal_length_cm'].isnull()) |
              (iris_data['petal_width_cm'].isnull())]
```

```python
iris_data.dropna(inplace=True)
```

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull()) |
              (iris_data['sepal_width_cm'].isnull()) |
              (iris_data['petal_length_cm'].isnull()) |
              (iris_data['petal_width_cm'].isnull())]
```

```python
iris_data.to_csv('iris-data-clean.csv', index=False)
iris_data_clean = pd.read_csv('iris-data-clean.csv')
```

```python
sb.pairplot(iris_data_clean, hue='class')
sb.plt.show()
```
