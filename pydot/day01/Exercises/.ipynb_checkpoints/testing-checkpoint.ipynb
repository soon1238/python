```python
students=[
    {
        'name':"soon",
        'marks':[55,65,75,95,100],
        'total_score':0,
        'avg':0.0
    },
    
    {
        'name':"tom",
        'marks':[55,65,73,95,60],
        'total_score':0,
        'avg':0.0
    },
    {
        'name':"ken",
        'marks':[55,66,78,88,100],
        'total_score':0,
        'avg':0.0
    }
]
    
for student in students:
    print("before calculation")
    print student
    total=0
    avg=0
#     print student["marks"]
    for mark in student["marks"]:
        total=total+mark
#     print total
    avg= total/len(student["marks"])
#     print ("average:%.2f" % avg)
    student['total_score']=total
    student['avg']=avg
    print ("***after calculation*** ")
    print student
    

        
```

```python
class Point():
    def __init__(self,x,y):
        self.x=x
        self.y=y
        self.name="soon"
    def describe(self):
        print("%d, %d, %s" %(self.x, self.y, self.name) )
    def add (self,p2):
        self.x+=p2.x
        self.y+=p2.y
        
p1=Point(10,20)
p2=Point (1,2)
p1.add(p2)
p1.describe()
```
