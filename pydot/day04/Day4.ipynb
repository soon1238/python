```python
import requests
import json
from pprint import pprint

def getdata(url,resp_type="text"):
    resp = requests.get(url)
    content=""
    if resp_type=="text":
        content=resp.text
    elif resp_type=="json":
        content=resp.json()
    else:
        pass
    return content

url="https://swapi.co/api/people/1"
resp=requests.get(url).json()
pprint(resp)
pprint(type(resp))



```
