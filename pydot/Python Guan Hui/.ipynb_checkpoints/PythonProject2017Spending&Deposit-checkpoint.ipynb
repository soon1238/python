```python

import csv
with open('2017Spending.csv','r') as csvfile:
    spendreader = csv.reader(csvfile)
    i = 0
    val = 0
    for f in spendreader: #print row by row in csv file.
        if (i == 0):
            i+=1
            continue
        print(f)
        
        val=val+float(f[1]) #adds up the spending with decimal places
    print ("Total spending for 2017 is $%.2f" %(val)) #keep the decimal place to 2

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('2017Spending.csv') #read file using pandas
# df.info() #print out the info
```

```python
df.info()
```

```python
df.plot() #line graph of daily spending. x-axis is before changing it to months.
plt.show()
```

```python
df.plot.bar() #bar graph
```

```python
plt.show()
```

```python
df1 = df.copy()
df1["Date"] = pd.to_datetime(df1["Date"]) #converting the date column to date values
```

```python
df1 #showing the data in a table
```

```python
per = df1.Date.dt.to_period("M") # grouping the daily spend into monthly spend
g = df1.groupby(per)

month_df = g.sum() #sum of spending per month

#Plotting
month_df.plot.bar()
plt.show()
```

```python
per = df1.Date.dt.to_period("M")
g = df1.groupby(per)

month_df = g.min() #min spend per month

#Plotting
month_df.plot.bar()
plt.show()
```

```python
per = df1.Date.dt.to_period("M")
g = df1.groupby(per)

month_df = g.max() #max spend per month

#Plotting
month_df.plot.bar()
plt.show()
```
