```python
from pprint import pprint
file_name = "icms_file.txt"

f = open(file_name, "r")

icms_code_dict_l4 = {}

for line in f:
    result = line.split('|')
    icms_code = result[0]
    desc = result[1]
    
    level_count = len(icms_code.split('.'))
#     print  ( icms_code , " ->  ", level_count,  desc )
    
    if level_count == 4:
#         print  ( icms_code , " ->  ", level_count,  desc )
        icms_code_dict_l4[icms_code] = {'desc': desc}
    
pprint ( icms_code_dict_l4 )
```
